```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies= movies.merge(credits,on='title')
```

```python
movies.info()
```

```python
#id
#geners
#overview
#cast
#crew
#keyword
#title
movies=movies[['movie_id','title','genres','overview','cast','crew','keywords']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval(movies.iloc[0].genres)      #ast.literal_eval use for covert string into list
```

```python
import ast
def covert(obj):
    L=[]
    for i in ast.literal_eval(obj):    #ast.literal_eval use for covert string into list
        L.append(i['name'])
    return L 
```

```python
movies['genres']
```

```python
movies['genres']=movies['genres'].apply(covert)
```

```python
movies['genres']
```

```python
movies['keywords']=movies['keywords'].apply(covert)
```

```python
def covert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):    #ast.literal_eval use for covert string into list
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(covert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags','genres']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y= []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out().tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectors)
```

```python
def recommend(movies):
    movies_index = new_df[new_df['title'] == movies].index[0]
    distances=similarity[movies_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
def filter_movies_by_genre(target_genre):
    """
    Filters movies by a specific genre.

    Parameters:
        movie_data (dict): A dictionary containing movie information with keys 'movie_id', 'title', 'tags', 'genres'.
        target_genre (str): The genre to filter movies by.

    Returns:
        list: A list of dictionaries with 'movie_id' and 'title' for movies that match the target genre.
    """
    if 'genres' not in new_df or 'title' not in new_df or 'movie_id' not in new_df:
        raise ValueError("Input data must contain 'movie_id', 'title', and 'genres' keys.")

    filtered_movies = []

    for movie_id, title, genres in zip(new_df['movie_id'], new_df['title'], new_df['genres']):
        if target_genre.lower() in [genre.lower() for genre in genres]:
            filtered_movies.append({
                'movie_id': movie_id,
                'title': title
            })

    return [movie['title'] for movie in filtered_movies][:10]

# Example usage
# filtered = filter_movies_by_genre(movie_data, "Comedy")
# print(filtered)
```
